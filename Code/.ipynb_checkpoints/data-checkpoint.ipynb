```python
data_patient = data.drop(['Hospital_code', 'Hospital_type_code', 'City_Code_Hospital',
                          'Hospital_region_code', 'City_Code_Patient'], axis = 1)
```

```python
data_patient.dtypes
```

```python
data_patient = func.get_day_visit(data_patient)
```

```python
data_patient
```

```python
data_patient['Ward_Type'] = pd.Categorical(data_patient['Ward_Type'])
data_patient['Ward_Type_code'] = data_patient['Ward_Type'].cat.codes

data_patient['Ward_Facility_Code'] = pd.Categorical(data_patient['Ward_Facility_Code'])
data_patient['Ward_Facility_Code_code'] = data_patient['Ward_Facility_Code'].cat.codes

data_patient.Department = pd.Categorical(data_patient.Department)
data_patient['Department_code'] = data_patient.Department.cat.codes

data_patient['Severity of Illness'] = pd.Categorical(data_patient['Severity of Illness'])
data_patient['Severity_of_Illness_code'] = data_patient['Severity of Illness'].cat.codes

data_patient['Type of Admission'] = pd.Categorical(data_patient['Type of Admission'])
data_patient['Type_of_Admission_code'] = data_patient['Type of Admission'].cat.codes

data_patient['Age'] = pd.Categorical(data_patient['Age'])
data_patient['Age_code'] = data_patient['Age'].cat.codes

data_patient['Stay'] = pd.Categorical(data_patient['Stay'])
data_patient['Stay_code'] = data_patient['Stay'].cat.codes
```

```python
categories_code = dict()
categories_code['Ward_Type'] = list(data_patient['Ward_Type'].cat.categories)
categories_code['Ward_Facility_Code'] = list(data_patient['Ward_Facility_Code'].cat.categories)
categories_code['Department'] = list(data_patient['Department'].cat.categories)
categories_code['Severity of Illness'] = list(data_patient['Severity of Illness'].cat.categories)
categories_code['Type of Admission'] = list(data_patient['Type of Admission'].cat.categories)
categories_code['Age'] = list(data_patient['Age'].cat.categories)
categories_code['Stay'] = list(data_patient['Stay'].cat.categories)
```

```python
with open('categories_code.pkl', 'wb') as file:
    pickle.dump(categories_code, file)
```
