```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import normalize

from lifelines import LogLogisticAFTFitter
from lifelines import LogNormalAFTFitter
from lifelines import CoxPHFitter, WeibullAFTFitter, WeibullFitter
from lifelines.calibration import survival_probability_calibration
```

```python
data = pd.read_csv('data_patient_num.csv')
data.dropna()
y = data['n_day_stay']
data = data.drop(['n_day_stay'], axis = 1)
data= data.apply(lambda x: (x - np.mean(x)) / (np.max(x) - np.min(x)))
# data['output'] = 1
data['n_day_stay'] = y
data = data.drop(['Stay_code'], axis=1)
data
```

```python
y = data['n_day_stay']
data_ml = data.drop(['case_id', 'patientid', 'Admission_Deposit'], axis = 1) 
# data_ml['Stay_code'] += 1
data_ml.head()
```

```python
X_train, X_test = train_test_split(data_ml, test_size=0.1, random_state=101)
```

## Cox's Model

```python
cph = CoxPHFitter()
cph.fit(X_train, duration_col='n_day_stay')
```

```python
cph.print_summary()
```

```python
cph.plot()
```

```python
predict = cph.predict_median(X_test)
predict
```

```python
err = abs(X_test.n_day_stay - predict)/(X_test.n_day_stay + 1) * 100
```

```python
err.value_counts()
```

```python
err.mean()
```

Err mean 
 **$48.5$**

## RandomForest

```python
y_train = X_train['n_day_stay']
X_train = X_train.drop(['n_day_stay'], axis = 1)
y_test = X_test['n_day_stay']
X_test = X_test.drop(['n_day_stay'], axis = 1)
```

```python
rf = RandomForestClassifier()
rf.fit(X_train, y_train)
```

```python
rf.score(X_test, y_test)
```

```python
predict = rf.predict(X_test)
```

```python
err = abs(y_test - predict)/(y_test + 1) * 100
err.mean()
```

```python
importances = pd.DataFrame({'feature':X_train.columns,'importance':np.round(rf.feature_importances_,3)})
importances = importances.sort_values('importance',ascending=False).set_index('feature')
print (importances)
importances.plot.bar()
```

## NN Model

```python
import tensorflow as tf
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Dense(units=5, activation='relu', dtype='float64'),
    tf.keras.layers.Dense(units=3, activation='relu', dtype='float64'),
    tf.keras.layers.Dense(201, dtype='float64'),
])
model.compile(optimizer='adam',
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])
```

```python
model.fit(X_train, y_train, epochs=10)
```

```python
predictions = model.predict(X_test)
```
